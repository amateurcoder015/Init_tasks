```python
rfm = merged_df.groupby('customer_id').agg({
    'timestamp': lambda x: (snapshot_date - x.max()).days,  # recency
    'transaction_id':  'count',                                    # frequency
    'revenue':          'sum'                                       # monetary
})
rfm.columns = ['Recency', 'Frequency', 'Monetary']
```

```python
rfm['R_bin'] = pd.qcut(rfm['Recency'], 3, labels=['High','Medium','Low'])
# recency is inverse logic  lower days = better  High

rfm['F_bin'] = pd.qcut(rfm['Frequency'], 3, labels=['Low','Medium','High'])
rfm['M_bin'] = pd.qcut(rfm['Monetary'], 3, labels=['Low','Medium','High'])
```

```python
def assign_segment(row):
    if row['R_bin']=='High' and row['F_bin']=='High' and row['M_bin']=='High':
        return 'VIP'
    elif row['R_bin']=='Low' and row['F_bin']=='Low':
        return 'Churn Risk'
    else:
        return 'Regular'

rfm['Segment'] = rfm.apply(assign_segment, axis=1)
```

```python
def find_suspicious_transactions(full_data):

    # Condition 1: quantity > 100 and price < 10
    cond_price_qty = (full_data['quantity'] > 100) & (full_data['price'] < 10)
    suspicious_qty_price = full_data.loc[cond_price_qty, ['transaction_id', 'customer_name']].copy()
    suspicious_qty_price['reason'] = 'High quantity & low price'

    # Condition 2: more than 3 purchases by same customer within same hour
    full_data['hour'] = full_data['timestamp'].dt.floor('H')  # round down to the hour

    # Count purchases per customer per hour
    purchase_counts = (
        full_data.groupby(['customer_name', 'hour'])['transaction_id']
        .transform('count')
    )
    cond_high_freq = purchase_counts > 3
    suspicious_high_freq = full_data.loc[cond_high_freq, ['transaction_id', 'customer_name']].copy()
    suspicious_high_freq['reason'] = 'More than 3 purchases in an hour'

    # Combine both suspicious sets
    suspicious_df = pd.concat([suspicious_qty_price, suspicious_high_freq], ignore_index=True)

    # Drop duplicates (in case a transaction fits both conditions)
    suspicious_df.drop_duplicates(subset='transaction_id', inplace=True)

    return suspicious_df
```

```python
def customer_lifetime_value(transactions_df, customer_id, discount_rate=0.1):
    """
    Calculate customer lifetime value (CLV) with time-discounting.
    
    CLV = (revenue  discount_factor^month_index)
    
    Parameters
    ----------
    transactions_df : pd.DataFrame
        DataFrame containing transaction data
        MUST have columns: [customer_id, transaction_date, revenue]
    
    customer_id : int / str
        Customer ID to calculate CLV for
    
    discount_rate : float
        Monthly discount rate (default 0.1)
    
    Returns
    -------
    float : Calculated CLV
    """
    
    # filter only this customer's transactions (vectorized mask)
    cust_tx = transactions_df.loc[transactions_df['customer_id'].eq(customer_id), 
                                  ['timestamp','revenue']]

    if cust_tx.empty:
        return 0.0  # customer not found
    
    # sort to ensure month order is correct
    cust_tx = cust_tx.sort_values('timestamp')
    
    # compute month index relative to first purchase (0,1,2,3,...)
    first_date = cust_tx['timestamp'].min()
    cust_tx['month_index'] = (
        cust_tx['timestamp'].dt.to_period('M').astype(int)
        - first_date.to_period('M').ordinal
)

    
    # discount factor = (1 / (1+rate)) ^ month_index   (vectorized)
    discount_factor = (1 / (1 + discount_rate)) ** cust_tx['month_index']
    
    # CLV =  revenue  discount factor  (vectorized multiply + sum)
    clv = (cust_tx['revenue'] * discount_factor).sum()
    
    return float(clv)
```
